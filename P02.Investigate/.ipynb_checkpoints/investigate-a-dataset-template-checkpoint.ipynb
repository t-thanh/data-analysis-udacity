### Research Question 1:  What is the correlation between: budget, revenue, voting and popularity?

```python
q1_df = data[['revenue_adj', 'budget_adj', 'popularity', 'vote_average']]
f1 = sns.pairplot(q1_df, kind="reg", diag_kind="kde", diag_kws=dict(shade=True))
f1.fig.suptitle('Correlation plots for budget_adj, revenue_adj, popularity and vote_average\n')
f1.fig.tight_layout(rect=[0, 0.03, 1, 0.95])
```

```python
corr = q1_df.corr()
_ , ax = plt.subplots( )
cmap = sns.diverging_palette( 240 , 10 , as_cmap = True )
_ = sns.heatmap(corr,cmap = cmap,square=True, cbar_kws={ 'shrink' : .9 }, ax=ax, annot = True, annot_kws = { 'fontsize' :12})
```

Adjusted revenue, adjusted budget, vote average and popularity have positive correlation with each other.

### Research Question 2: Evolution of run time and voting over the years?

```python
# Create a subset with averages grouped by year
avg_set = data.groupby('release_year', as_index=False)['runtime','vote_average'].mean()
# Linear regression on average movie runtimes over the years
g = sns.jointplot(x="release_year", y="runtime", data=avg_set, space=0.5, kind="reg", ylim=(90, 130), xlim=(1959, 2016), color="b", height=8)
plt.title('Linear Regression of Average Runtime over the Years', fontdict={'fontsize': 18})
plt.xlabel('Year', fontdict={'fontsize': 16})
plt.ylabel('Runtime', fontdict={'fontsize': 16});
```

```python
g = sns.jointplot(x="release_year", y="vote_average", data=avg_set, space=0.5, kind="reg", ylim=(5.7, 7), xlim=(1959, 2016), height=8)
plt.title('Evolution of Average Runtime over the Years', fontdict={'fontsize': 18})
plt.xlabel('Year', fontdict={'fontsize': 16})
plt.ylabel('Vote', fontdict={'fontsize': 16});
```

Run time and average vote is reduced over time.

### Research Question 3: Who is the actor can assure the profit of movies?

```python
#make a subset from 
def count_subset(x):
    #concatenate all the rows of the genrs.
    data_plot = data[x].str.cat(sep = '|')
    data_ = pd.Series(data_plot.split('|'))
    #conts each of the genre and return.
    info = data_.value_counts(ascending=False)
    return info
#count number of movies done by each actor.
count_actor = count_subset('cast')

#plot the 20 most popular actor plot.
count_actor.iloc[:20].plot.bar(figsize=(13,6),fontsize=12)

#setup the title and the labels of the plot.
plt.title("Most Frequent Actor",fontsize=15)
plt.xticks(rotation = 70)
plt.xlabel('Actor',fontsize=13)
plt.ylabel("Number Of Movies",fontsize= 13)
```

```python
# Split the dataframe by genre
actor = data['cast'].str.split('|').apply(pd.Series).stack()
actor.index = actor.index.droplevel(-1)
actor.name = 'actor'
df_by_actor = data.join(actor)
df_by_actor['count'] = df_by_actor.groupby('actor')['actor'].transform('count')
df_by_actor.head()
```

```python
actor_top20_by_freq = df_by_actor.groupby(['count','actor'])['vote_average'].mean()
df_actor_top20_by_freq = actor_top20_by_freq.to_frame('vote_average').reset_index().sort_values(['count'], ascending=[False])
df_actor_top20_by_freq.iloc[:20]
plt.figure(figsize=(13,6))
chart = sns.scatterplot(x='actor', y='count', size='vote_average', data=df_actor_top20_by_freq.iloc[:20], sizes=(100, 1000))
plt.xticks(rotation=90)
```
